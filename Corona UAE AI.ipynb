# AI Time

```python
def do_func(days,cases):
    #making the model

    import numpy as np
    import time
    daysformodel=np.asarray(days).reshape(-1,1)
    t0=time.time()
    model.fit(daysformodel,cases) #Training the model
    print("Training time:",time.time()-t0)
    print("Evaluating...")
    print("Accuracy:",model.score(daysformodel,cases))
    print("Predicting...")
    preds=model.predict(daysformodel)
    return preds
```

```python
def graphprint(days,cases,preds):
    plt.plot(days,cases,color='blue')
    plt.plot(days,preds,color='red')
    plt.ylabel("Number of cases")
    plt.xlabel("Days")
    plt.show()
```

```python
import numpy as np
def mean_absolute_error(y_true, y_pred):
    error = 0
    for yt, yp in zip(y_true, y_pred):
        error += np.abs(yt - yp)
    return error / len(y_true)

def mean_percentage_error(y_true, y_pred):
    error = 0
    for yt, yp in zip(y_true, y_pred):
        error += (yt - yp) / yt
    return error / len(y_true)
```

```python
def errors(cases,preds):
    from sklearn import metrics
    print("MSE: ",end="")
    print(metrics.mean_squared_error(cases,preds))
    print("Explained Variance: ",end="")
    print(metrics.explained_variance_score(cases,preds))
    print("R^2 error: ",end="")
    print(metrics.r2_score(cases,preds))
    print("MAE: ",end="")
    print(mean_absolute_error(cases,preds))
    print("MPE: ",end="")
    print(mean_percentage_error(cases,preds))
```
